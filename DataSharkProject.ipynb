# **Renderizar el DataFrame con Pandas游냪**

```python

df = pd.read_csv('world_population.csv')
num_filas = len(df)

num_columnas = len(df.columns)
print("N칰mero de filas:", num_filas)
print("N칰mero de columnas:", num_columnas)

df
```

```python
import pandas as pd
import pycountry
def get_alpha_3(location):
  try:
    return pycountry.countries.get(name=location).alpha_3
  except:
    return None
df['Code']=df['Country/Territory'].apply(lambda x: get_alpha_3(x))
print(df.head(6))
```

```python
import plotly.express as px
figura=px.choropleth(df,locations='Code',
                     color='Area (km)',hover_name='Country/Territory',
                     color_continuous_scale=px.colors.sequential.Plasma)
figura.show()
```

```python
# Crear una figura y ejes para los subgr치ficos
fig, axs = plt.subplots(8, 1, figsize=(20, 80), sharex=True)

# Lista de a침os
years = ['1970', '1980', '1990', '2000', '2010', '2015', '2020', '2022']

# Iterar sobre cada a침o y crear un gr치fico para ese a침o
for i, year in enumerate(years):
    # Seleccionar los datos correspondientes al a침o
    df_year = df_europa[['Country/Territory', f'{year} Population']]
    df_year_sorted = df_year.sort_values(by=f'{year} Population', ascending=False)

    # Crear el gr치fico de barras para el a침o actual
    axs[i].barh(df_year_sorted['Country/Territory'], df_year_sorted[f'{year} Population'], color='skyblue')
    axs[i].set_xlabel(f'Poblaci칩n en {year}')
    axs[i].set_ylabel('Pa칤s')
    axs[i].set_title(f'Poblaci칩n total de cada pa칤s en Europa en {year}')
    axs[i].invert_yaxis()  # Invertir el eje y para ordenar de mayor a menor

# Ajustar el espaciado entre los subgr치ficos
plt.tight_layout()

# Mostrar la figura
plt.show()
```
